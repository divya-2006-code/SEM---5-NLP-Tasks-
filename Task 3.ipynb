```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt


# News Articles
articles = [
    "Cybersecurity experts are developing advanced methods to prevent data breaches",
    "Cloud computing platforms are helping businesses scale their digital services",
    "Data science and analytics are driving smarter business decisions worldwide",
    "Cybersecurity and cloud technologies are becoming essential for modern enterprises"
]

# 1. Convert text to lowercase
text = " ".join(articles).lower()

# 2. Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Remove stopwords
stop_words = set(stopwords.words('english'))

# 3. Tokenization
tokens = word_tokenize(text)

filtered_tokens = [
    word for word in tokens
    if word not in stop_words
]

print("Tokens After Cleaning:\n")
print(filtered_tokens)

# 4. Word Frequency Distribution
word_freq = Counter(filtered_tokens)

print("\nWord Frequency Distribution:\n")
print(word_freq)

# 5. Vocabulary
vocabulary = set(filtered_tokens)

print("\nVocabulary:\n")
print(vocabulary)

print("\nVocabulary Size:", len(vocabulary))

# 6. Top 5 Most Frequent Words
top5 = word_freq.most_common(5)

print("\nTop 5 Frequent Words:")
for word, freq in top5:
    print(word, ":", freq)

# 7. Type Token Ratio (TTR)
total_tokens = len(filtered_tokens)
unique_tokens = len(vocabulary)

ttr = unique_tokens / total_tokens

print("\nType-Token Ratio (TTR):", round(ttr,3))

# 8. Dominant Topic Analysis
topics = {
    "Cybersecurity": ["cybersecurity", "breaches", "experts"],
    "Cloud Computing": ["cloud", "computing", "platforms"],
    "Data Science": ["data", "science", "analytics"],
    "Business": ["business", "businesses", "enterprises"]
}

topic_scores = {}

for topic, keywords in topics.items():
    score = sum(word_freq[word] for word in keywords)
    topic_scores[topic] = score

print("\nTopic Scores:")
print(topic_scores)

dominant_topic = max(topic_scores, key=topic_scores.get)

print("\nDominant Topic:", dominant_topic)

# 9. Bar Chart
words = [word for word, freq in top5]
freqs = [freq for word, freq in top5]

plt.figure(figsize=(8,5))
plt.bar(words, freqs)

plt.title("Top 5 Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")

plt.show()
```
